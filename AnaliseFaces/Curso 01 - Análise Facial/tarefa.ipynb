```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn
```

```python
imagem = cv2.imread('imagens/menina.png')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(imagem)
plt.title("Garota")
```

```python
roi = imagem[200:400, 500:700]
```

```python
plt.figure(figsize=(20,10))
plt.imshow(roi)
plt.title("Corpo inteiro")
```

```python
roi_gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(20,10))
plt.imshow(roi_gray, cmap="gray")
plt.title("Corpo inteiro")
```

Para finalizar, precisamos inserir este recorte exatamente na imagem original, no mesmo lugar.

Lembre-se que a imagem precisa estar colorida, com isso, o recorte poder ser inserido. Tambm precisamos convert-la em 3 canais de cores: vermelho, verde e azul (do ingls Red, Green e Blue).

A primeira parte  converter a regio de interesse para RGB. Note que a simples converso no vai colorir a imagem novamente. Na imagem em escala de cinza havia somente um canal representando a intensidade de branco ou preto. Agora, esse valor vai ser replicado nos outros trs canais (RGB). Visualmente, no vai mudar absolutamente nada.

```python
roi_rgb = cv2.cvtColor(roi_gray, cv2.COLOR_GRAY2RGB)
```

```python
imagem[200:400, 500:700] = roi_rgb
```

```python
plt.figure(figsize=(20,10))
plt.imshow(imagem)
plt.title("Corpo inteiro")
```
