```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("imagens/px-girl.jpg")
```

```python
plt.imshow(imagem)
```

Essa imagem no foi feita dessa forma aqui, tem uma distoro de cores. Por que aconteceu isso? Porque o OpenCV, por padro, abre as imagens no formato .bgr, que so as iniciais para Blue, Green e Red, que seriam azul, verde e vermelho.

Se formos usar o pyplot, ele usa um outro padro, o padro RGB. Nesse espao de cores, ele s inverte o R pelo B, que  o Red, Blue e Green, ento precisamos fazer uma converso antes de plotar igual plotamos aqui.

```python
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
imagem_rgb.shape
```

Nem sempre precisamos das cores, pois elas aumentam muito a quantidade de clculos necessrios

```python
imagem_gray = cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(imagem_gray, cmap="gray")
```

```python
imagem_gray.shape
```

Vamos separar a regio de interesse de uma imagem, muito til pra focar no que exatamente  importante na anlise

```python
imagem = cv2.imread("imagens/px-people.jpg")
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
imagem_rgb.shape
```

```python
imagem_roi = imagem_rgb[100:200, 1050:1200]
plt.imshow(imagem_roi)
```

Vamos salvar essa imagem em um arquivo externo.

```python
cv2.imwrite("imagens/imagem_roi.png", imagem_roi)
```

Mas tem um probleminha: se vocs verem, essa imagem est um pouquinho distorcida. Por que disso? Porque quando eu mandei escrever, escrevemos a imagem no RGB, s que por padro, o OpenCV usa o BGR.

Para escrevermos sem ter distores, vamos fazer a converso antes, ento fica imagem_roi_bgr = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR) - ento tudo o que for comando original do OpenCV, como o imread, imwrite, sempre trabalhamos com BGR. Agora eu vou chamar o mesmo comando cv2.imwirte() - vou escrever no mesmo arquivo - ("imagem_roi.png", ) e vou passar aqui a varivel , imagem_roi_bgr).

```python
imagem_roi_bgr = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
cv2.imwrite("imagens/imagem_roi.png", imagem_roi_bgr)
```
