```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("imagens/px-girl.jpg")
```

```python
plt.imshow(imagem)
```

Essa imagem não foi feita dessa forma aqui, tem uma distorção de cores. Por que aconteceu isso? Porque o OpenCV, por padrão, abre as imagens no formato “.bgr”, que são as iniciais para Blue, Green e Red, que seriam azul, verde e vermelho.

Se formos usar o “pyplot”, ele usa um outro padrão, o padrão “RGB”. Nesse espaço de cores, ele só inverte o “R” pelo “B”, que é o “Red”, “Blue” e “Green”, então precisamos fazer uma conversão antes de plotar igual plotamos aqui.

```python
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
imagem_rgb.shape
```

Nem sempre precisamos das cores, pois elas aumentam muito a quantidade de cálculos necessários

```python
imagem_gray = cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(imagem_gray, cmap="gray")
```

```python
imagem_gray.shape
```

Vamos separar a região de interesse de uma imagem, muito útil pra focar no que exatamente é importante na análise

```python
imagem = cv2.imread("imagens/px-people.jpg")
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
imagem_rgb.shape
```

```python
imagem_roi = imagem_rgb[100:200, 1050:1200]
plt.imshow(imagem_roi)
```

Vamos salvar essa imagem em um arquivo externo.

```python
cv2.imwrite("imagens/imagem_roi.png", imagem_roi)
```

Mas tem um probleminha: se vocês verem, essa imagem está um pouquinho distorcida. Por que disso? Porque quando eu mandei escrever, escrevemos a imagem no “RGB”, só que por padrão, o OpenCV usa o “BGR”.

Para escrevermos sem ter distorções, vamos fazer a conversão antes, então fica “imagem_roi_bgr = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)” - então tudo o que for comando original do OpenCV, como o “imread”, “imwrite”, sempre trabalhamos com “BGR”. Agora eu vou chamar o mesmo comando “cv2.imwirte()” - vou escrever no mesmo arquivo - “("imagem_roi.png", )” e vou passar aqui a variável “, imagem_roi_bgr)”.

```python
imagem_roi_bgr = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
cv2.imwrite("imagens/imagem_roi.png", imagem_roi_bgr)
```
